```python
# Libs
# Libs para modelagem e matrizes
import numpy as np
import pandas as pd

# Libs para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Libs para ignorar avisos
import warnings

# Desabilitar avisos
warnings.filterwarnings("ignore")
```

```python
# ANÁLISE DAS AÇÕES DA MAGALU
# Lendo os dados
base_dados = pd.read_excel("Vase_004 - Magalu - Sem Resolução.xlsx")

# Verificando
base_dados.head()
```

```python
base_dados.shape
```

```python
base_dados.info()
```

```python
base_dados.describe()
```

```python
# Séries Temporais
# tenho que colocar a data como index

dados = base_dados.set_index("Data")
dados.head()
```

```python
plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(16,5))
plt.title("Análise das ações da magalu - Fechamento", fontsize=14, loc="left")
plt.plot(dados.index, dados["Fechamento"]);

plt.xlabel("Período da Cotação")
plt.ylabel("Valor da Ação  (R$)");
```

```python
# Verifica os 5 últimos dados do banco de dados
dados.tail()
```

```python
# Fazendo média móvel
# .rolling(5) - está agrupando os valores de 5 em 5
# ele vai retornar a média da quantidade de valores que eu colocar em rolling
Media_Movel = dados["Fechamento"].rolling(5).mean()
Media_Tendencia = dados["Fechamento"].rolling(30).mean()

plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(16,5))
plt.title("Análise das Ações da magalu - Fechamento", fontsize=15, loc="left")
plt.plot(dados.index, dados["Fechamento"])
plt.plot(dados.index, Media_Movel)
plt.plot(dados.index, Media_Tendencia)

plt.xlabel("Período da Cotação")
plt.ylabel("Valor da Ação (R$)");
```

```python
sns.boxplot(data=dados, x="Fechamento")
```

```python
# Box plot mensal para poder analisar com esse tipo grafico contando com uma
# bbase de dados dia a dia

base_dados["Mes"] = base_dados["Data"].dt.month

base_dados.head()
```

```python
plt.figure(figsize=(16,5))
sns.boxplot(data=base_dados, x="Mes", y="Fechamento");
```

```python
base_dados.groupby(by=["Mes"]).describe()["Fechamento"]
```

```python
#
grafico = go.Figure(
    data=[
        go.Candlestick(
            x=dados.index,
            open = dados["Abertura"],
            high = dados["Maior"],
            low = dados["Menor"],
            close = dados["Fechamento"],
        )
    ]
)

grafico.update_layout(xaxis_rangeslider_visible=False) # Retira a barra inferior do gráfico

grafico.show()
```

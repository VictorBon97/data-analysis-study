```python
# Bibliotecas

# Modelagem e Matríz
import numpy as np
import pandas as pd

# Análises Gráficas
import matplotlib.pyplot as plt
import seaborn  as sns

# Ignorar Avisos
import warnings

# Desativar Avisos
warnings.filterwarnings("ignore")
```

```python
base_dados = pd.read_csv("Salary_Data.csv")

base_dados.shape
```

```python
base_dados.head()
```

```python
base_dados.isnull().sum()
```

```python
base_dados.info()
```

```python
# Renomeando as Colunas

base_dados.rename(columns={
    "YearsExperience":"xp",
    "Salary":"renda"
},inplace=True)
```

```python
# Gráfico dos Nulos

sns.heatmap(base_dados.isnull(), cbar=False)
```

```python
base_dados.describe()
```

```python
# kdeplot serve para ver a densidade das informações
plt.figure(figsize=(16,5))
plt.title("Densidade da Renda")

sns.kdeplot(base_dados["renda"], color="green");
```

```python
plt.figure(figsize=(16,5))
plt.title("Densidade da XP")

sns.kdeplot(base_dados["xp"]);
```

```python
# Distribuição
plt.figure(figsize=(5,5))
sns.boxplot(base_dados["renda"]);
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(base_dados["xp"])
```

```python
# Distribuição
plt.figure(figsize=(15,6))
sns.scatterplot(data=base_dados, x="renda", y="xp");
```

```python
plt.figure(figsize=(15,6))
sns.regplot(data=base_dados, x="renda", y="xp")
```

```python
# Correlação perason
plt.figure(figsize=(16,10))
correlacao = base_dados.corr("pearson")
sns.heatmap(correlacao, annot=True, )
```

```python
# Correlação spearman
plt.figure(figsize=(16,10))
correlacao = base_dados.corr("spearman")
sns.heatmap(correlacao, annot=True, )
```
